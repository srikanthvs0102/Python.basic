```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Company_Financials_India.csv')
```

```python
df.describe()
```

```python
sns.lineplot(x='Year', y='Net_Profit_INR_Crores', data=df)

plt.title("Net Profit Over Time")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Revenue_INR_Crores'], marker='o')

plt.title("Revenue Growth Trend")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```

```python
sns.barplot(x='Year', y='Capex_INR_Crores', data=df)

plt.title("Capital Expenditure per Year")
plt.xticks(rotation=45)
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.scatterplot(x='EPS_INR',
                y='Dividend_per_Share_INR',
                data=df)

plt.title("EPS vs Dividend Relationship")
plt.show()
```
