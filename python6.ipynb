```python
#price distribution
plt.Figure(figsize=(8,5))
sns.histplot(df['price'],kde=True)
plt.title('laptop_price_dataset_v4.csv')
plt.show()

```

```python
#price By brand
plt.figure(figsize=(10,5))
sns.boxplot(x='brand',y='price',data=df)
plt.title('laptop_price brand')
plt.xticks(rotation=45)
plt.show()
```

```python
#Correlation Analysis
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('correlation matrix')
plt.show()
```

```python
#price using cpu
plt.figure(figsize=(10,5))  
sns.boxplot(x='cpu',y='price',data=df)
plt.title('laptop_price cpu')
plt.xticks(rotation=45)
plt.show()  
```

#Which brand sells the most expensive laptops?

```python
#price more expensive brand
plt.figure(figsize=(10,5))
sns.boxplot(x='brand',y='price',data=df)
plt.title('laptop_price brand')
plt.xticks(rotation=45)
plt.show()
```

#Do gaming laptops cost more than student laptops?


```python
#price comparison of gaming vs non gaming laptops
plt.figure(figsize=(10,5))
sns.boxplot(x='brand',y='price',data=df)
plt.title('laptop_price gaming')    
plt.xticks(rotation=45)
plt.show()
```

#Does more RAM significantly increase price?

```python
#price distribution by RAM
plt.figure(figsize=(10,5))  
sns.boxplot(x='ram',y='price',data=df)
plt.title('laptop_price ram')
plt.xticks(rotation=45)
plt.show()
```

#Do dedicated GPUs increase laptop price?

```python
#price using gpu
plt.figure(figsize=(10,5))      
sns.boxplot(x='brand',y='price',data=df)
plt.title('laptop_price gpu')
plt.xticks(rotation=45)
plt.show()
```

#Which CPU category has highest average price?

```python
#Average price by CPU
avg_price_cpu=df.groupby('cpu')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=avg_price_cpu.index,y=avg_price_cpu.values)
plt.title('Average price by CPU')
plt.xticks(rotation=45)
plt.show()
```

KEY INSIGHTS:

1.First laptop prices are compared.

2.Then they are listed on The highest to lowest order

3.Laptop price is compared using RAM,CPU, and dedicated GPU

4.Price comparison between Gaming and Non-Gaming laptops has been done.

5.Laptops price is compared using their brands 

6.Laptops average and highest prices are compared using Power of CPU

Conclusion:

The analysis clearly shows that laptop pricing is driven by a combination of hardware specifications and brand positioning. Among all factors, CPU, RAM, and storage type (SSD) have the strongest impact on price, while brand value adds an additional premium layer.
